```python
import pandas as pd 
import numpy as np
```

```python
resultados_spre= pd.read_pickle("resultados1.pkl")
resultados_cpre=pd.read_pickle("resultados2.pkl")
```

```python
m_escolhidos=['RFC','BGT','XGB','LGB','RNE']
```

```python
precisao= pd.DataFrame( columns=['modelo','pre-processamento','p50','p40','p30','p20','p10','p1'])
for i in m_escolhidos:
    p50=round(np.nanmean(resultados_spre.loc[(resultados_spre['modelo']==i) & (resultados_spre['prob']==0.5),'Precision']),3)
    p40=round(np.nanmean(resultados_spre.loc[(resultados_spre['modelo']==i) & (resultados_spre['prob']==0.4),'Precision']),3)
    p30=round(np.nanmean(resultados_spre.loc[(resultados_spre['modelo']==i) & (resultados_spre['prob']==0.3),'Precision']),3)
    p20=round(np.nanmean(resultados_spre.loc[(resultados_spre['modelo']==i) & (resultados_spre['prob']==0.2),'Precision']),3)
    p10=round(np.nanmean(resultados_spre.loc[(resultados_spre['modelo']==i) & (resultados_spre['prob']==0.1),'Precision']),3)
    p1=round(np.nanmean(resultados_spre.loc[(resultados_spre['modelo']==i) & (resultados_spre['prob']==0.01),'Precision']),3)

    precisao.loc[len(precisao)]=[i,'nao',p50,p40,p30,p20,p10,p1]

    p50=round(np.nanmean(resultados_cpre.loc[(resultados_cpre['modelo']==i) & (resultados_cpre['prob']==0.5),'Precision']),3)
    p40=round(np.nanmean(resultados_cpre.loc[(resultados_cpre['modelo']==i) & (resultados_cpre['prob']==0.4),'Precision']),3)
    p30=round(np.nanmean(resultados_cpre.loc[(resultados_cpre['modelo']==i) & (resultados_cpre['prob']==0.3),'Precision']),3)
    p20=round(np.nanmean(resultados_cpre.loc[(resultados_cpre['modelo']==i) & (resultados_cpre['prob']==0.2),'Precision']),3)
    p10=round(np.nanmean(resultados_cpre.loc[(resultados_cpre['modelo']==i) & (resultados_cpre['prob']==0.1),'Precision']),3)
    p1=round(np.nanmean(resultados_cpre.loc[(resultados_cpre['modelo']==i) & (resultados_cpre['prob']==0.01),'Precision']),3)

    precisao.loc[len(precisao)]=[i,'sim',p50,p40,p30,p20,p10,p1]
```

```python
recall= pd.DataFrame( columns=['modelo','pre-processamento','p50','p40','p30','p20','p10','p1'])
for i in m_escolhidos:
    p50=round(np.nanmean(resultados_spre.loc[(resultados_spre['modelo']==i) & (resultados_spre['prob']==0.5),'Recall']),3)
    p40=round(np.nanmean(resultados_spre.loc[(resultados_spre['modelo']==i) & (resultados_spre['prob']==0.4),'Recall']),3)
    p30=round(np.nanmean(resultados_spre.loc[(resultados_spre['modelo']==i) & (resultados_spre['prob']==0.3),'Recall']),3)
    p20=round(np.nanmean(resultados_spre.loc[(resultados_spre['modelo']==i) & (resultados_spre['prob']==0.2),'Recall']),3)
    p10=round(np.nanmean(resultados_spre.loc[(resultados_spre['modelo']==i) & (resultados_spre['prob']==0.1),'Recall']),3)
    p1=round(np.nanmean(resultados_spre.loc[(resultados_spre['modelo']==i) & (resultados_spre['prob']==0.01),'Recall']),3)

    recall.loc[len(recall)]=[i,'nao',p50,p40,p30,p20,p10,p1]

    p50=round(np.nanmean(resultados_cpre.loc[(resultados_cpre['modelo']==i) & (resultados_cpre['prob']==0.5),'Recall']),3)
    p40=round(np.nanmean(resultados_cpre.loc[(resultados_cpre['modelo']==i) & (resultados_cpre['prob']==0.4),'Recall']),3)
    p30=round(np.nanmean(resultados_cpre.loc[(resultados_cpre['modelo']==i) & (resultados_cpre['prob']==0.3),'Recall']),3)
    p20=round(np.nanmean(resultados_cpre.loc[(resultados_cpre['modelo']==i) & (resultados_cpre['prob']==0.2),'Recall']),3)
    p10=round(np.nanmean(resultados_cpre.loc[(resultados_cpre['modelo']==i) & (resultados_cpre['prob']==0.1),'Recall']),3)
    p1=round(np.nanmean(resultados_cpre.loc[(resultados_cpre['modelo']==i) & (resultados_cpre['prob']==0.01),'Recall']),3)

    recall.loc[len(recall)]=[i,'sim',p50,p40,p30,p20,p10,p1]
```
