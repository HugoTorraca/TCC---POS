```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("dados_2.csv",sep=";")
```

Gerando graficos para analisar os dados

```python
verao_0=data[(data["Incosistente"] == 0) & (data["Estacao"] == 1)]
verao_1=data[(data["Incosistente"] == 1) & (data["Estacao"] == 1)]
outono_0=data[(data["Incosistente"] == 0) & (data["Estacao"] == 2)]
outono_1=data[(data["Incosistente"] == 1) & (data["Estacao"] == 2)]
inverno_0=data[(data["Incosistente"] == 0) & (data["Estacao"] == 3)]
inverno_1=data[(data["Incosistente"] == 1) & (data["Estacao"] == 3)]
primavera_0=data[(data["Incosistente"] == 0) & (data["Estacao"] == 4)]
primavera_1=data[(data["Incosistente"] == 1) & (data["Estacao"] == 4)]
```

```python
maximo=round((max(verao_0['Valor'].max(),verao_1['Valor'].max(),verao_0['m_postos'].max(),verao_1['m_postos'].max())/1000),2)*1000 +20
plt.figure(figsize=(6, 4))

plt.scatter(verao_0['Valor'], verao_0['m_postos'], marker='.', color='b', alpha=0.8, label='Ver√£o Consistente')
plt.scatter(verao_1['Valor'], verao_1['m_postos'], marker='.', color='r', alpha=0.8, label='Ver√£o Inconsistente')

plt.xlabel("PSAT");
plt.ylabel("M√©dia Postos");
plt.xlim(0,maximo);
plt.ylim(-0.5,maximo);
plt.title(" Psat x M√©dia dos Postos")
plt.legend(loc='best');
```

```python
maximo=round((max(outono_0['Valor'].max(),outono_1['Valor'].max(),outono_0['m_postos'].max(),outono_1['m_postos'].max())/1000),2)*1000 +20
plt.figure(figsize=(6, 4))

plt.scatter(outono_0['Valor'], outono_0['m_postos'], marker='x', color='b', alpha=0.8, label='Outono Consistente')
plt.scatter(outono_1['Valor'], outono_1['m_postos'], marker='x', color='r', alpha=0.8, label='Outono Inconsistente')

plt.xlabel("PSAT");
plt.ylabel("M√©dia Postos");
plt.xlim(0,maximo);
plt.ylim(-1,maximo);
plt.title(" Psat x M√©dia dos Postos")
plt.legend(loc='best');
```

```python
maximo=round((max(inverno_0['Valor'].max(),inverno_1['Valor'].max(),inverno_0['m_postos'].max(),inverno_1['m_postos'].max())/1000),2)*1000 +20
plt.figure(figsize=(6, 4))

plt.scatter(inverno_0['Valor'], inverno_0['m_postos'], marker='s', color='b', alpha=0.8, label='Inverno Consistente')
plt.scatter(inverno_1['Valor'], inverno_1['m_postos'], marker='s', color='r', alpha=0.8, label='Inverno Inconsistente')

plt.xlabel("PSAT");
plt.ylabel("M√©dia Postos");
plt.xlim(0,maximo);
plt.ylim(-1,maximo);
plt.title(" Psat x M√©dia dos Postos")
plt.legend(loc='best');
```

```python
maximo=round((max(primavera_0['Valor'].max(),primavera_1['Valor'].max(),primavera_0['m_postos'].max(),primavera_1['m_postos'].max())/1000),2)*1000 +20
plt.figure(figsize=(6, 4))

plt.scatter(primavera_0['Valor'], primavera_0['m_postos'], marker='^', color='b', alpha=0.8, label='Primavera Consistente')
plt.scatter(primavera_1['Valor'], primavera_1['m_postos'], marker='^', color='r', alpha=0.8, label='Primavera Inconsistente')

plt.xlabel("PSAT");
plt.ylabel("M√©dia Postos");
plt.xlim(0,maximo);
plt.ylim(-1,maximo);
plt.title(" Psat x M√©dia dos Postos")
plt.legend(loc='best');
```
